```python
import re

fcn = re.compile("^extern ([^ ]+) ASAP3_EXPORT CALL_CONV ([^(]+)\(([^)]+)\);", flags=re.MULTILINE)

import os
header_path ="CANapAPI.h"

with open(header_path, "r") as fid:
    header = fid.read()

import ctypes

lookup = {
    "unsigned long": "ctypes.c_ulong",
    "bool": "ctypes.c_bool",
    "unsigned short": "ctypes.c_ushort",
    "unsigned int": "ctypes.c_uint",
    "UINT": "ctypes.c_uint",
    "unsigned char": "ctypes.c_ubyte",
}

def to_c_type(arg_type):
    if arg_type.startswith("T"):
        return arg_type
    if arg_type in [
        "version_t",
        "Appversion",
        "DWORD",
        "e_RamMode",
        "DBObjectInfo",
        "tDriverType",
        "tFifoSize",
        "MeasurementListEntries",
        "tMeasurementState",
        "EnRecorderState",
        "SecProfileEntry"
    ]:
        return arg_type
    if arg_type in lookup:
        return lookup[arg_type]
    if hasattr(ctypes, f"c_{arg_type}"):
        return f"ctypes.c_{arg_type}"
    raise Exception(arg_type)

for f in fcn.findall(header):
    f = [g.strip() for g in f]
    return_type, function_name, function_args = f
    
    if "Init" not in function_name:
        continue

    print(f"# {function_name}")
    print(f"dll.{function_name}.argtypes=(")
    for arg in function_args.split(","):
        arg_words = arg.strip().split(" ")
        var_name = arg_words[-1]
        arg_type = " ".join(arg_words[:-1]).strip()
        pointer = False
        if arg_type.endswith("*"):
            arg_type=arg_type.rstrip("*").strip()
            pointer=True
        elif var_name.startswith("*"):
            var_name=var_name.lstrip("*").strip()
            pointer=True
        if arg_type == "const char":
            c_type = "ctypes.c_char_p"
        else:
            c_type = to_c_type(arg_type)
            if pointer:
                c_type=f"ctypes.POINTER({c_type})"
        print(f"""    {c_type},""")
    print(")")
    c_type = to_c_type(return_type)

    print(f"dll.{function_name}.restype={c_type}")
    print()
```
