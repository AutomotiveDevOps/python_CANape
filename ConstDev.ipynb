```python
import ctypes
import os
import re
```

```python
with open(os.path.abspath("CANapAPI.h"), "r") as fid:
    CANapAPI_h = fid.read()
```

```python
const_re = re.compile("#define ([\w_]+)\s+([\d]+)\s+//!< ([\w\s]+)")
```

```python
for const in const_re.finditer(CANapAPI_h):
    break
```

```python
name = const.group(1).strip()
value = int(const.group(2))
comment = const.group(3).strip()
```

```python
print(f"""# {comment}
{name}={value}
""")
```

```python
for const in const_re.finditer(CANapAPI_h):
    name = const.group(1).strip()
    value = int(const.group(2))
    comment = const.group(3).strip()
    print(f"""# {comment}
{name}={value}
""")
```
